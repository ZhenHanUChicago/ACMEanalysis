```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0015.1294_0015.1295_0015.1296_0015.1297.csv")
value = "omega_NE"
```

```python
# reduced chi2
uncertainty = "uncertainty_" + value
mean_value = (df[value] * df[uncertainty]**(-2)).sum() / np.sum(df[uncertainty]**(-2))
photon_errorbar = np.sum(df[uncertainty]**(-2))**(-0.5)
reduced_chi2 = ((df[value] - mean_value)**2 * df[uncertainty]**(-2)).sum() / (len(df) - 1)
print("Mean value:", mean_value)
print("Photon errorbar:", photon_errorbar)
print("Reduced chi2:", reduced_chi2)
print("scaled errorbar:", photon_errorbar * np.sqrt(reduced_chi2))
```
