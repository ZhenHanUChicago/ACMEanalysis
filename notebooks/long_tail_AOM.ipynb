```python
import pandas as pd
import plotly.express as px
import numpy as np
df = pd.read_csv(r"C:\Users\ACME\Downloads\RigolDS4.csv")
```

```python
y = df['CH1V'].to_numpy().reshape((-1,20000)).mean(axis = 0)
baseline = y[10000:15000].mean()
y = y - baseline
y = np.convolve(y, np.ones(50)/50, mode='valid')
y = np.log10(y)
x = np.arange(len(y))*8E-10
px.line(x=x , y = y)
```
