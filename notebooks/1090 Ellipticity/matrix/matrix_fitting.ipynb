```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
def weighted_mean(value, uncertainty,group):
    value = group[value]
    uncertainty = group[uncertainty]
    weights = 1 / uncertainty**2
    return np.sum(value * weights) / np.sum(weights)
```

```python
df_1090Ellip_phase_prepon = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0016.0830_0016.0831_0016.0832_0016.0833_0016.0834_.csv")
df_1090HWP_phase_prepon = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0016.0843_0016.0844_0016.0845_0016.0846_0016.0847_.csv")

df_1090Ellip_polarimeter_prepoff = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0016.0795.csv")
df_1090HWP_polarimeter_prepoff = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0016.0797.csv")
df_1090Ellip_phase_prepoff = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0016.0799.csv")
df_1090HWP_phase_prepoff = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0016.0801.csv")
```

```python
phi_NE_prepoff = np.concatenate([
    df_1090HWP_phase_prepoff.sort_values(by = "STIRAP_1090_HWP_Dither")["phi_NE"].to_numpy(),
df_1090Ellip_phase_prepoff.sort_values(by = "1090_STIRAP_ellipticity_angle")["phi_NE"].to_numpy()
])

Ellipticity_NE_prepoff = np.concatenate([
    df_1090HWP_polarimeter_prepoff.sort_values(by = "STIRAP_1090_HWP_Dither")["Am_NE"].to_numpy()/2,
df_1090Ellip_polarimeter_prepoff.sort_values(by = "1090_STIRAP_ellipticity_angle")["Am_NE"].to_numpy()/2
])

phi_NE_prepon = np.concatenate([
    df_1090HWP_phase_prepon.sort_values(by = "STIRAP_1090_HWP_Dither").groupby("STIRAP_1090_HWP_Dither").apply(lambda group: weighted_mean("phi_NE", "uncertainty_phi_NE", group)).to_numpy(),
    df_1090Ellip_phase_prepon.sort_values(by = "1090_STIRAP_ellipticity_angle").groupby("1090_STIRAP_ellipticity_angle").apply(lambda group: weighted_mean("phi_NE", "uncertainty_phi_NE", group)).to_numpy()
])
```

```python
# phi_NE_prepon = beta0 + beta1 * phi_NE_prepoff + beta2 * Ellipticity_NE_prepoff
X = np.column_stack((phi_NE_prepoff, Ellipticity_NE_prepoff))
X = sm.add_constant(X)  # Adds a constant term to the predictor
model = sm.OLS(phi_NE_prepon, X).fit()
print(model.summary())
print("Fitted parameters (beta0, beta1, beta2):", model.params)
```

```python
phi_E_prepoff = np.concatenate([
    df_1090HWP_phase_prepoff.sort_values(by = "STIRAP_1090_HWP_Dither")["phi_E"].to_numpy(),
df_1090Ellip_phase_prepoff.sort_values(by = "1090_STIRAP_ellipticity_angle")["phi_E"].to_numpy()
])

Ellipticity_E_prepoff = np.concatenate([
    df_1090HWP_polarimeter_prepoff.sort_values(by = "STIRAP_1090_HWP_Dither")["Am_E"].to_numpy()/2,
df_1090Ellip_polarimeter_prepoff.sort_values(by = "1090_STIRAP_ellipticity_angle")["Am_E"].to_numpy()/2
])

phi_E_prepon = np.concatenate([
    df_1090HWP_phase_prepon.sort_values(by = "STIRAP_1090_HWP_Dither").groupby("STIRAP_1090_HWP_Dither")["phi_E"].mean().to_numpy(),
    df_1090Ellip_phase_prepon.sort_values(by = "1090_STIRAP_ellipticity_angle").groupby("1090_STIRAP_ellipticity_angle")["phi_E"].mean().to_numpy(),
])
```

```python
# phi_E_prepon = beta0 + beta1 * phi_E_prepoff + beta2 * Ellipticity_E_prepoff
X = np.column_stack((phi_E_prepoff, Ellipticity_E_prepoff))
X = sm.add_constant(X)  # Adds a constant term to the predictor
model = sm.OLS(phi_E_prepon, X).fit()
print(model.summary())
print("Fitted parameters (beta0, beta1, beta2):", model.params)
```
