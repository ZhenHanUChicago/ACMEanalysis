```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0015.2858_0015.2859_0015.2860_0015.2861_0015.2862_.csv")
df1 = df1[df1['Block_Readoutkz-1'] == 1]
df2 = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0015.2985_0015.2986_0015.2987_0015.2988_0015.2989_.csv")
df2 = df2[df2['Block_Readoutkz-1'] == 1]
df3 = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0015.3134_0015.3135_0015.3136_0015.3137_0015.3138_.csv")
df3 = df3[df3['Block_Readoutkz-1'] == 1]
df4 = pd.read_csv(r"C:\ACME_analysis\multiple_results\sequencedf_result\0015.3247_0015.3248_0015.3249_0015.3250_0015.3251_.csv")
df4 = df4[df4['Block_Readoutkz-1'] == 1]

df = pd.concat([df1,df2,df3,df4],ignore_index=True)
```

```python
df['omega_NE'].hist(bins = 50)
df['uncertainty_omega_NE']
```

```python
omega_NE = df['omega_NE']
uncertainty_omega_NE = df['uncertainty_omega_NE']
weighted_mean_omega_NE = np.average(omega_NE, weights=1/uncertainty_omega_NE**2)
print(weighted_mean_omega_NE)
```

```python
# combined error bar
weighted_std_omega_NE = np.sqrt(1/np.sum(1/uncertainty_omega_NE**2))
print(weighted_std_omega_NE)
```
