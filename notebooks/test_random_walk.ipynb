```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df = pd.DataFrame({'x': np.random.randn(1000)})
df['s'] = df['x'].expanding().mean()
plt.plot(df['s'])
plt.ylim(-0.2,0.2)
# plot the 2sigma bounds
plt.fill_between(df.index, -1/np.sqrt(df.index+1), 1/np.sqrt(df.index+1), color='gray', alpha=0.5)
```
